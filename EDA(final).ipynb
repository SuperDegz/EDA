```python
def display_stats(sp, column):
    if column in sp.columns:
        # Convert the column to numeric, coercing errors to NaN
        sp[column] = pd.to_numeric(sp[column], errors='coerce')
        
        stats = sp[column].dropna().agg(['mean', 'median', 'std'])
        
        print(f"\nStatistics for streams column:")
        print("=" * 40)
        print(f"{'Metric':<25}{'Value'}")
        print("=" * 40)
        for metric, value in stats.items():
            print(f"{metric.capitalize() + ':' :<25}{value:.2f}")
        print("=" * 40)
    else:
        print(f"The '{column}' column does not exist in the dataset.")


display_stats(sp, 'streams')
```

```python
sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

variables = ['released_year', 'artist_count']
titles = ['Released Year', 'Artist Count']

for i, var in enumerate(variables):
    sns.histplot(sp[var], bins=30, kde=True, ax=axes[i, 0], color='blue' if var == 'released_year' else 'orange')
    axes[i, 0].set_title(f'Distribution of {titles[i]}')
    axes[i, 0].set_xlabel(titles[i])
    axes[i, 0].set_ylabel('Frequency')


    sns.boxplot(x=sp[var], ax=axes[i, 1], color='lightblue' if var == 'released_year' else 'lightyellow')
    axes[i, 1].set_title(f'Box Plot of {titles[i]}')
    axes[i, 1].set_xlabel(titles[i])


plt.tight_layout()
plt.show()
```

```python
def check_non_numeric(column_name):
    non_numeric = sp[~sp[column_name].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
    if not non_numeric.empty:
        print(f"Non-numeric values found in '{column_name}':")
        print(non_numeric)

check_non_numeric('streams')
check_non_numeric('bpm')
check_non_numeric('danceability_%')
check_non_numeric('energy_%')

sp.loc[:, 'streams'] = pd.to_numeric(sp['streams'], errors='coerce')
sp.loc[:, 'bpm'] = pd.to_numeric(sp['bpm'], errors='coerce')
sp.loc[:, 'danceability_%'] = pd.to_numeric(sp['danceability_%'], errors='coerce')
sp.loc[:, 'energy_%'] = pd.to_numeric(sp['energy_%'], errors='coerce')


correlations_streams = sp[['streams', 'bpm', 'danceability_%', 'energy_%']].corr()
print("\nCorrelation with streams:")
print(correlations_streams['streams'])


cor_dance_energy = sp['danceability_%'].corr(sp['energy_%'])
print("\nCorrelation between danceability_% and energy_%:", cor_dance_energy)

cor_valence_acoustic = sp['valence_%'].corr(sp['acousticness_%'])
print("Correlation between valence_% and acousticness_%:", cor_valence_acoustic)
```

```python
spotify_tracks = sp['in_spotify_playlists'].sum()
deezer_tracks = sp['in_deezer_playlists'].sum()
apple_tracks = sp['in_apple_playlists'].sum()

popular_spotify = sp[sp['in_spotify_charts'] > 80].shape[0]
popular_deezer = sp[sp['in_deezer_charts'] > 80].shape[0]
popular_apple = sp[sp['in_apple_charts'] > 80].shape[0]

platforms = ['Spotify', 'Deezer', 'Apple']
total_tracks = [spotify_tracks, deezer_tracks, apple_tracks]
popular_tracks = [popular_spotify, popular_deezer, popular_apple]

fig, ax = plt.subplots(figsize=(8, 6))

bar_width = 0.35
index = range(len(platforms))

bar1 = ax.bar(index, total_tracks, bar_width, label='Total Tracks')
bar2 = ax.bar([i + bar_width for i in index], popular_tracks, bar_width, label='Popular Tracks')

ax.set_xlabel('Platform')
ax.set_ylabel('Number of Tracks')
ax.set_title('Comparison of Tracks and Popular Tracks by Platform')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(platforms)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
sp.loc[:, 'streams'] = pd.to_numeric(sp['streams'], errors='coerce')
sp.loc[:, 'bpm'] = pd.to_numeric(sp['bpm'], errors='coerce')
sp.loc[:, 'danceability_%'] = pd.to_numeric(sp['danceability_%'], errors='coerce')
sp.loc[:, 'energy_%'] = pd.to_numeric(sp['energy_%'], errors='coerce')
sp.loc[:, 'valence_%'] = pd.to_numeric(sp['valence_%'], errors='coerce')
sp.loc[:, 'acousticness_%'] = pd.to_numeric(sp['acousticness_%'], errors='coerce')

average_streams_by_bpm = sp.groupby('bpm')['streams'].mean().reset_index()
average_streams_by_danceability = sp.groupby('danceability_%')['streams'].mean().reset_index()

print("Average streams by BPM:")
print(average_streams_by_bpm)

print("\nAverage streams by Danceability:")
print(average_streams_by_danceability)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(average_streams_by_bpm['bpm'], average_streams_by_bpm['streams'], marker='o')
plt.title('Average Streams by BPM')
plt.xlabel('BPM')
plt.ylabel('Average Streams')

plt.subplot(1, 2, 2)
plt.plot(average_streams_by_danceability['danceability_%'], average_streams_by_danceability['streams'], marker='o', color='purple')
plt.title('Average Streams by Danceability')
plt.xlabel('Danceability (%)')
plt.ylabel('Average Streams')

plt.tight_layout()
plt.show()
```
