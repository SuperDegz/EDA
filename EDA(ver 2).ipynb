```python
def display_stats(sp, column):
    if column in sp.columns:
        # Convert the column to numeric, coercing errors to NaN
        sp[column] = pd.to_numeric(sp[column], errors='coerce')
        
        # Drop NaN values
        stats = sp[column].dropna().agg(['mean', 'median', 'std'])
        
        print(f"\nStatistics for streams column:")
        print("=" * 40)
        print(f"{'Metric':<25}{'Value'}")
        print("=" * 40)
        for metric, value in stats.items():
            print(f"{metric.capitalize() + ':' :<25}{value:.2f}")
        print("=" * 40)
    else:
        print(f"The '{column}' column does not exist in the dataset.")

# Call the function with the DataFrame and the column name
display_stats(sp, 'streams')
```

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Define the variables and titles for easy iteration
variables = ['released_year', 'artist_count']
titles = ['Released Year', 'Artist Count']

for i, var in enumerate(variables):
    # Histogram
    sns.histplot(sp[var], bins=30, kde=True, ax=axes[i, 0], color='blue' if var == 'released_year' else 'orange')
    axes[i, 0].set_title(f'Distribution of {titles[i]}')
    axes[i, 0].set_xlabel(titles[i])
    axes[i, 0].set_ylabel('Frequency')

    # Box plot
    sns.boxplot(x=sp[var], ax=axes[i, 1], color='lightblue' if var == 'released_year' else 'lightyellow')
    axes[i, 1].set_title(f'Box Plot of {titles[i]}')
    axes[i, 1].set_xlabel(titles[i])

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
# Assuming the dataset has columns 'track_name' and 'streams'
top_tracks = sp[['track_name', 'streams']].nlargest(5, 'streams')

# Reset index for better formatting
top_tracks.reset_index(drop=True, inplace=True)

# Display the top 5 most streamed tracks in a presentable format
print("Top 5 Most Streamed Tracks:")
print(top_tracks.to_string(index=False))
```
